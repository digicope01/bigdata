### 배열의 인덱싱과 슬라이싱

```python
# 1차원 배열의 인덱싱과 슬라이싱 : list와 동일한 방법
d = np.arange(12)
print(d)
print(d[0],d[1],d[-1])  # 인덱싱
print(d[:-1],d[::3])
```

```python
# 2차원 배열의 인덱싱과 슬라이싱
# data[행,열]
d = np.arange(12).reshape(3,4)
print(d)
# 인덱싱
print(d[0,0])  # 0
print(d[0][0]) # 0
print(d[2,2])  # 10

# 슬라이싱
print(d[0])  
print(d[0,:])    # print(d[0]) 와 같은 결과
print(d[:,:-1])  # 마지막열을 제외
print(d[::2,::2])
d[0] = 20
print(d)
d[::2,::3] = 30
print(d)
```

```python
# 3차원 배열의 인덱싱과 슬라이싱
# data[면,행,렬]
d = np.arange(12).reshape(2,2,3)
print(d)
# 인덱싱
print(d[1,0,1])  # 7
print(d[0,1,2 ]) # 5
print(d[1][0][1]) # old style
# 슬라이싱
print(d[:,:-1,:-1]) # (2,1,2)
print(d[:,1:,1:])   # (2,1,2)
print(d[:,:,:-1])   # (2,2,2)
print(d[:,:,-1])    # (2,2)
print(d[:,-1,-1]    # (2,)
```

### 배열의 형상 다루기: 편평화(flatten), 전치(transpose),reshape

```python
# flatten() : 다차원 배열을 1차원으로 만든다
d = np.arange(12).reshape(3,4)
print(d)
f = d.flatten()
print(f)  # 1차원 배열 (12,) 
print(d.reshape(12)) # 동일한 결과

# 데이터 순서를 거꾸로 변경
print(d[::-1])   # 행의 순서를 거꾸로 변경
print(d[:,::-1]) # 열의 순서를 거꾸로 변경
print(d[::-1,::-1])# 행과 열의 순서를 모두 거꾸로 변경

# transpose() : 전치 행렬, 행과 열을 서로 맞바꾸는 함수
t = d.transpose()
print(t)
print(t.shape)  # (3,4)  --> (4,3)

d.flat = 0  # 배열의 모든 값을 특정값으로 변경
print(d)
```

### numpy 속성(Attribute) : 클래스의 인스턴스 멤버 변수
- T,  flat,  ndim,  nbytes,  size,  dtype

```python
a = np.arange(12).reshape(3,4)
print(a)
print(a.shape)
print(a.nbytes)   # 48
print(a.size)     # 12
print(a.dtype)    # int32
print(a.ndim)     # 2 (차원)
print(a.T)
a.flat = -1
print(a)
```

```python
# numpy 통계 함수들
a = np.arange(12).reshape(3,4)
print(a)
print('Max:',np.max(a))
print('Min:',np.min(a))
print('Mean:',np.mean(a))
print('Sum:',np.sum(a))
print('Std:',np.std(a))
print('Median:',np.median(a))
print('Var:',np.var(a))

from scipy.stats import scoreatpercentile
print('Score at percentile 50:',scoreatpercentile(a,50))
print('Score at percentile 25:',scoreatpercentile(a,25))
```

```python
# raccoon (오리너구리사진 처리)
# View : 원본 객체와 메모리를 공유하는 사본
# Copy : 원본객체와 다른 별도의 메모리를 사용하느 사본

import scipy.misc
import matplotlib.pyplot as plt

face = scipy.misc.face() # read only ndarray
face2 = face.copy()   # 복사본 생성: read and write 가능
print(type(face))
print(face.shape) # (768, 1024, 3)
print(face.dtype) # uint8
print(face.nbytes)
print(face.size)
print(face.ndim)

f_view = face2.view()
f_copy = face2.copy()

# f_view.flat = 0
f_copy.flat = 0

plt.subplot(221)
plt.imshow(face2)

plt.subplot(222)
plt.imshow(f_view)

plt.subplot(223)
plt.imshow(f_copy)

plt.show()


```

 ### 배열 합치기
 - 수직(v,행)으로 합치기 : vstack(), row_stack(), concatenate(.,axis = 0)
 - 수평(h,열)으로 합치기 : hstack(), column_stack(), concatenate(.,axis = 1)
 

```python
# 수직으로 합치기
a = np.arange(9).reshape(3,3)
b = a * 2
print(a)
print(b)
r = np.vstack((a,b))
print(r)
c = np.row_stack((a,b))
print(c)
x = np.concatenate((a,b),axis = 0) # 행
print(x)
```

```python
# 수평으로 합치기
a = np.arange(9).reshape(3,3)
b = a * 2
print(a)
print(b)
r = np.hstack((a,b))
print(r)
c = np.column_stack((a,b))
print(c)
x = np.concatenate((a,b),axis = 1) # 열
print(x)
```

### 쪼개기
- vsplit(배열, 행의 갯수) : 수직 , 행,  split(,,axis = 0)
- hsplit(배열, 열의 갯수) : 수평 , 열,  split(,,axis = 1)

```python
a = np.arange(12).reshape(4,3)
print(a)

print(np.vsplit(a,4))
print(np.split(a,4,axis = 0))

a = np.arange(12).reshape(3,4)
print(a)

print(np.hsplit(a,4))
print(np.split(a,4,axis = 1))
```

## 행렬(Matrix) : 선형대수
### 역행렬구하기


```python
A = np.mat("2 4 6;4 2 6;10 -4 18")
print(type(A))
print(A)
# B = np.array([[ 2,  4 , 6 ],
#  [ 4,  2 , 6],
#  [10, -4, 18 ]])
# print(type(B))
# print(B)

inverse = np.linalg.inv(A)   # A의 역행렬 반환
print(inverse)
print(A*inverse)
print(np.eye(3))
print(A*inverse - np.eye(3))
```

```python
# 선형방정식 풀이 : 선형시스템을 해석
A = np.mat("1 -2 1;0 2 -8;-4 5 9")
print("A:",A)
b = np.array([0,8,-9])
print("b:",b)
# 1*x0 - 2*x1 + 1*x2 = 0
# 0*x0 + 2*x1 - 8*x2 = 8
# -4*x0 + 5*x1 + 9*x2 = -9
```

```python
x = np.linalg.solve(A,b)
print('Solution:',x)
print('Check\n',np.dot(A,x))
```

###  2차원 배열의 원소 접근
- 배열명[[행1,행2,행3, ...,행n],[열1,열2,열3, ... , 열n]]

```python
a2 = np.arange(10,100,10).reshape(3,3)
print(a2)
print(a2[[0,2],[0,1]])  # 2차원 배열의 원소 접근
print(a2[0,0],a2[2,1])  # 배열명[행번호,열번호]

print(a2[[0,2]])
print(a2[[0,1,2]])
print(a2[[0,1]])
print(a2[[0]])
```
